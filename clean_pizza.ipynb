```python
import pandas as pd
import numpy as np


df=pd.read_csv('businesses_reviews.csv')

# Assuming 'df' is your DataFrame and 'business_id' is the column with business IDs
df = df.drop_duplicates(subset='business_id', keep='first')
```

```python
# Function to extract attribute options
def extract_columns(row):
    try:
        data_dict = eval(row['attributes'])
        return data_dict
    except:
        return {}

# Apply the function to create new columns
df = pd.concat([df, df.apply(extract_columns, axis=1, result_type='expand')], axis=1)



# Function to extract meal options
def extract_meal_options(row):
    try:
        meal_options_dict = eval(row['GoodForMeal'].replace("'", '"'))
        return pd.Series(meal_options_dict)
    except:
        return pd.Series({})

# Apply the function to create new columns
meal_options = df.apply(extract_meal_options, axis=1)

# Add the new columns to the DataFrame
df = pd.concat([df, meal_options], axis=1)




# Function to extract start and end times for each day
# Function to extract columns from the JSON-like string
def extract_hours(row):
    try:
        data_dict = eval(row['hours'])
        return data_dict
    except:
        return {}

# Apply the function to create new columns
df = pd.concat([df, df.apply(extract_hours, axis=1, result_type='expand')], axis=1)



# Function to extract parking options
def extract_parking_options(row):
    try:
        parking_options_dict = eval(row['BusinessParking'].replace("'", '"'))
        return pd.Series(parking_options_dict)
    except:
        return pd.Series({})

# Apply the function to create new columns
parking_options = df.apply(extract_parking_options, axis=1)

# Add the new columns to the DataFrame
df = pd.concat([df, parking_options], axis=1)


df.head()
```

```python
to_drop=['attributes', 'Ambience', 'hours', 'GoodForMeal', 'BusinessParking',
         'Unnamed: 0','useful', 'stars_y','categories','date','is_open','funny','cool',
         'RestaurantsTableService','HappyHour','DogsAllowed','BusinessAcceptsBitcoin','BYOB','Corkage',
        'ByAppointmentOnly','BYOBCorkage','DriveThru','Open24Hours','Smoking', 'categories','stars_y','breakfast']

df = df.drop(to_drop, axis=1)
df.head()
```

```python
# Remove 'u' and quotes from values
df['RestaurantsAttire'] = df['RestaurantsAttire'].str.replace("u'casual'", "casual")
df['RestaurantsAttire'] = df['RestaurantsAttire'].str.replace("u'formal'", "formal")
df['RestaurantsAttire'] = df['RestaurantsAttire'].str.replace("'", '')


df['Alcohol']=df['Alcohol'].str.replace("u'none'","None")
df['Alcohol']=df['Alcohol'].str.replace("u'beer_and_wine'","beer_and_wine")
df['Alcohol']=df['Alcohol'].str.replace("u'full_bar'","full_bar")
df['Alcohol']=df['Alcohol'].str.replace("'none'","None")
df['Alcohol']=df['Alcohol'].str.replace("'beer_and_wine'","beer_and_wine")
df['Alcohol']=df['Alcohol'].str.replace("'full_bar'","full_bar")


df['NoiseLevel']=df['NoiseLevel'].str.replace("u'average'", "average")     
df['NoiseLevel']=df['NoiseLevel'].str.replace("'average'","average")      
df['NoiseLevel']=df['NoiseLevel'].str.replace("u'quiet'","quiet")        
df['NoiseLevel']=df['NoiseLevel'].str.replace("u'loud'","loud")          
df['NoiseLevel']=df['NoiseLevel'].str.replace("'quiet'","quiet")         
df['NoiseLevel']=df['NoiseLevel'].str.replace("u'very_loud'","very_loud")      
df['NoiseLevel']=df['NoiseLevel'].str.replace("'loud'","loud")           
df['NoiseLevel']=df['NoiseLevel'].str.replace("'very_loud'","very_loud")


df['WiFi']=df['WiFi'].str.replace("u'no'","no")     
df['WiFi']=df['WiFi'].str.replace("u'free'","free")    
df['WiFi']=df['WiFi'].str.replace("'no'","no")        
df['WiFi']=df['WiFi'].str.replace("'free'","free")      
df['WiFi']=df['WiFi'].str.replace("u'paid'","paid")      
df['WiFi']=df['WiFi'].str.replace("'paid'","paid")   
```
